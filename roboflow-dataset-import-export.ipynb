```python
import yaml
import os
import shutil
from roboflow import Roboflow

with open("settings.yaml", "r") as f:
    config = yaml.safe_load(f)

rf = Roboflow(api_key=config["ROBOFLOW_API_KEY"])
project = rf.workspace(config["ROBOFLOW_WORKSPACE"]).project(config["ROBOFLOW_PROJECT"])
version = project.version(config["DATASET_VERSION"])
dataset = version.download("yolov8")

source_path=f"./{config['ROBOFLOW_PROJECT']}-{config['DATASET_VERSION']}"
destination_path=f"./datasets/{config['DATASET_TYPE']}"
shutil.move(f"./{config['ROBOFLOW_PROJECT']}-{config["DATASET_VERSION"]}", f"./datasets/{config['DATASET_TYPE']}")
#!mv ./{confi["ROBOFLOW_PROJECT"]}-* datasets/{os.environ["BUILDING_COMPONENT"]}
#!chmod -R 777 datasets/{os.environ["BUILDING_COMPONENT"]}
```
