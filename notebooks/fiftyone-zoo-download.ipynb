```python
import fiftyone
```

```python
# set max samples and dataset name 

max_samples=1000
dataset_name = "der_merger" + (f"-{max_samples}" if max_samples else "")

try:
    fiftyone.delete_dataset(dataset_name)
    print(f"Deleted existing dataset: {dataset_name}")
except:
    print(f"No existing dataset named: {dataset_name}")
```

```python
dataset = fiftyone.zoo.load_zoo_dataset(
    "coco-2017",
    splits=["train", "validation", "test"],
    label_types=["detections"],
    classes=["chair"],
    max_samples=max_samples,
)

# Rename the dataset
dataset.name = dataset_name
dataset.save()
```

```python
session = fiftyone.launch_app(dataset)
```
