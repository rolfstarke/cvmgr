# export for training with yolo

```python
# %%script echo "Skipping this cell"

#  select to models parent folder
datasets_parent_dir = pathlib.Path.cwd().parent / "datasets"

# choose a name for the new merged dataset
export_dataset = fiftyone.load_dataset("open-images-v7-train-validation-test")

splits = ["train", "test", "val"]

export_dir = datasets_parent_dir / "multiclass" / "open-image-v7-yolov5"
```

# some tag renaming and splits redistribution

```python

# Find all samples tagged with "validation" and retag them as "val"
validation_samples = export_dataset.match_tags("validation")
validation_samples.tag_samples("val")
validation_samples.untag_samples("validation")


print(f"Splits before redistribution: {export_dataset.count_sample_tags()}")
export_dataset.untag_samples(splits)
fiftyone.utils.random.random_split(export_dataset, {"train": 0.7, "test": 0.2, "val": 0.1})
print(f"Splits after redistribution: {export_dataset.count_sample_tags()}")
```

```python
# %%script echo "Skipping this cell"

# Export the splits
for split in splits:
    split_view = export_dataset.match_tags(split)
    split_view.export(
        export_dir=str(export_dir),
        dataset_type=fiftyone.types.YOLOv5Dataset,
        label_field="ground_truth",
        # overwrite=True,
        split=split,
        classes=settings["fiftyone_zoo"]["open-images-v7"]["classes"],
        progress=True,
    )
```
