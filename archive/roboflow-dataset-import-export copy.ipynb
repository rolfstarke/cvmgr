```python
import shutil
import subprocess
from roboflow import Roboflow


rf = Roboflow(api_key=ROBOFLOW_API_KEY)
project = rf.workspace(ROBOFLOW_WORKSPACE).project(ROBOFLOW_PROJECT)
version = project.version(DATASET_VERSION)
dataset = version.download("coco")


shutil.copytree(source_path, destination_path, dirs_exist_ok=True)
shutil.rmtree(source_path)

subprocess.run(["chmod", "-R", "777", f"{destination_path}"])

print(f"dataset copied to {destination_path} and permissions set to 777")
```

# convert from data.yaml to classes.txt

```python
import yaml

data_yaml = f"{destination_path}/data.yaml"
with open(data_yaml) as f:
    data = yaml.safe_load(f)
with open(f"{destination_path}/classes.txt", "w") as f:
    for name in data["names"]:
        f.write(f"{name}\n")
```
