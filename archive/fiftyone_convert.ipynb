import fiftyone  
# Convert a COCO detection dataset to CVAT image format

fiftyone convert \
    --input-dir /mnt/EBI-SHARE/06_Data/labelstudio/datasets/evaluation/room312_and_leuthener-2-coco \
    --input-type fiftyone.types.COCODetectionDataset \
    --output-dir /mnt/EBI-SHARE/06_Data/labelstudio/datasets/evaluation/room312_and_leuthener-2-cvat \
    --output-type fiftyone.types.CVATImageDataset

```python
import fiftyone as fo

dataset = fo.Dataset.from_dir(
    dataset_dir="/mnt/EBI-SHARE/06_Data/labelstudio/datasets/evaluation/room312_and_leuthener-2",
    dataset_type=fo.types.YOLOv5Dataset
)
dataset.export(
    export_dir="/mnt/EBI-SHARE/06_Data/labelstudio/datasets/evaluation/room312_and_leuthener-2-cvat",
    dataset_type=fo.types.CVATImageDataset
)
```
