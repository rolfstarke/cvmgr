```python
from ultralytics import YOLO
import yaml
from datetime import datetime

timestamp = datetime.now().strftime("%Y_%m_%d_%H_%M_%S")

with open("settings.yaml", "r") as f:
    config = yaml.safe_load(f)

model = YOLO("yolo11n.pt") 

datasets = config['DATASET_LOCATIONS']
training_configurations = config['TRAINING_CONFIGURATIONS']

for dataset in datasets:
    for configuration in training_configurations:
        # Only include keys that exist in configuration
        train_kwargs = {
            "data": dataset,
            "project": f"~/onemanstreasure/models/{config['DATASET_TYPE']}",
            "name": f"{timestamp}_model"
        }
        # Optional keys
        for key in ["time", "imgsz", "patience", "epochs", "multi_scale", "batch"]:
            if key in configuration:
                train_kwargs[key] = configuration[key]
        train_results = model.train(**train_kwargs)

# for dataset in datasets:
#     for configuration in training_configurations:
#         train_results = model.train(
#             data=dataset,  # Path to dataset configuration file
#             time=configuration["time"],  # maximum training time in hours
#             imgsz=configuration["imgsz"],  # Image size for, auto resizing images, effects accuracy - how needs to be evaluated
#             patience=configuration["patience"],
#             epochs=configuration["epochs"],  # Number of epochs to train
#             project=f"~/onemanstreasure/models/{config['DATASET_TYPE']}",
#             name=f"{timestamp}_model",
#             multi_scale=configuration["multi_scale"], # Optional: artificially variing imagesize for better generalization
#             batch=configuration("batch"),  # Use the default "chunks" to split the dataset for computation, -1 is auto 60% GPU utilization
#         )



#     print(f"Training on dataset: {dataset}")                   
# train_results = model.train(
#     data=os.environ["DATASET_LOCATION_LOCAL"]+"/data.yaml",  # Path to dataset configuration file
#     time=3, # maximum training time in hours
#     # epochs=1,
#     imgsz=640,  # Image size for, auto resizing images, effects accuracy - how needs to be evaluated
#     patience=15,
#     # batch=-1, # Use the default "cunks" to split the dataset for computation, -1 is auto 60% GPU utilization
#     # maybe batch will cause problems with multiple notebooks running
#     project=f"~/onemanstreasure/models/{config["DATASET_TYPE"]}"
#     name=f"{timestamp}_model",
#     # multi_scale=True, # Optional: artificially variing imagesize for better generalization
    
# )



```
